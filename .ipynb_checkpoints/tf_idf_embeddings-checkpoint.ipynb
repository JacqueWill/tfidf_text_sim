```python
def convert_lower_case(df):
    df = df.apply(lambda x: x.astype(str).str.lower())
    return df

def remove_punctuation(df):
    symbols = "!\"#$%&()*+-./:;<=''>?@[\]^_`{|}~\n½²⅓⅛¾⅔¼"
    for i in range(len(df)):
        text = df['text'][i]
        new_text = ""
        for words in text:
            if words not in symbols:
                new_text = new_text + words
        df['text'][i] = new_text
    return df

def remove_apos(df):
    return np.char.replace(df,"'","")

def remove_single_char(df):
    for i in range(len(df)):
        text = df["text"][i].split(" ")
        new_text = ""
        for word in text:
            if len(word)>1:
                new_text = new_text + " " + word
        df["text"][i] = new_text
    return df

def convert_numbers(df):
    for i in range(len(df)):
        text = df["text"][i].split(" ")
        new_text = ""
        for word in text:
            if word.isnumeric():
                new_text = new_text + " " + num2words(word)
            else:
                new_text = new_text + " " + word
        df["text"][i] = new_text
    return df

def remove_stop_words(df):
    for i in range(len(df)):
        text = df["text"][i].split(" ")
        new_text = ""
        for word in text:
            if word not in stop_words:
                new_text = new_text + " " + word
        df["text"][i] = new_text
    return df

def stemming(df):
    ps = PorterStemmer()
    for i in range(len(df)):
        text = df["text"][i].split(" ")
        new_text = ""
        for word in text:
            new_text = new_text + " " + ps.stem(word)
        df['text'][i] = new_text
    return df
        
```

```python
def preprocessing(df):
    df = convert_lower_case(df)
    df = remove_punctuation(df)
    df = remove_single_char(df)
#     print(df['text'][1])
#    df = convert_numbers(df)
#    print(df['text'][1])
    df = remove_stop_words(df)
    df = stemming(df)
    df = remove_punctuation(df)
#    df = convert_numbers(df)
    return df
```

```python
df = pd.read_csv("DBPEDIA_train.csv")
```

```python
df = preprocessing(df)
```

```python
# Calculating document frequency
def doc_freq(text):
    doc_fq={}
    for words in text:
        if words not in doc_fq:
            doc_fq[words] = 0
        doc_fq[words] += 1
    return doc_fq
```

```python
# Defining variables for TF-IDF
def calculate_tf(text, doc_fq):
    tf = {}
    
    word_count = len(text)
    for word, count in doc_fq.items():
        tf[word] = count / float(word_count)
        
    return tf

def calculate_idf(df_fqs):
    import math
    
    n = len(df_fqs)
    idf = {}
    
    for i in range(n):
        for word, count in df_fqs[i].items():
            if count > 0:
                try:
                    idf[word] += 1
                except:
                    idf[word] = 1
                
    for word, count in idf.items():
        idf[word] = (n/float(count))
        
    return idf

def calculate_tfidf(tf, idfs):
    tfidf = {}
    
    for word, val in tf.items():
        tfidf[word] = val * idfs[word]
        
    return tfidf
```

```python
#Calculate TF
tf_dicts = {}
for i in range(len(df)):
    tokens = df['text'][i].split(" ")
    try:
        tokens.remove(" ")
    except:
        pass
    doc_fq = doc_freq(tokens)
    tf_dicts[i] = calculate_tf(tokens, doc_fq)
    
```

```python
# Creating df_freq
df_fqs = {}
for i in range(len(df)):
    tokens = df["text"][i].split(" ")
    df_fqs[i] = doc_freq(tokens)
    
#Calculate IDF
idfs = calculate_idf(df_fqs)

#Calculate TF-IDF
tfidfs = {}
for i in range(len(df)):
    tfidf = calculate_tfidf(tf_dicts[i],idfs)
    tfidfs[i] = tfidf

```
